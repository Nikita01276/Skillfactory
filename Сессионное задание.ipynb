Задание 1 (10 баллов)
Обратный порядок слов в блоках текста

```python
# Открываем файл 
input_file = open('input.txt', encoding='utf-8')

# Читаем все строки
all_lines = []
line = input_file.readline()

while line:
    if line.strip():  # если строка не пустая
        all_lines.append(line.strip())
    line = input_file.readline()

input_file.close()

# Обрабатываем текст
result_lines = []
i = 0
block_size = 2

while i < len(all_lines):
    # Обрабатываем блок
    j = 0
    while j < block_size and i < len(all_lines):
        sentence = all_lines[i]
        # Разбиваем на слова
        words = sentence.split(' ')
        # Переворачиваем слова
        reversed_words = []
        k = len(words) - 1
        while k >= 0:
            reversed_words.append(words[k])
            k -= 1
        
        # Собираем обратно в строчку
        reversed_sentence = ' '.join(reversed_words)
        result_lines.append(reversed_sentence)
        i += 1
        j += 1
    
    # Добавляем пустую строку между блоками
    result_lines.append('')
output_file = open('output.txt', 'w', encoding='utf-8')

for line in result_lines:
    output_file.write(line + '\n')
    print(line)
output_file.close()
```

Задание 2 (6 баллов)
Сжатие строки

```python
def compress_string(s):
    # Если строка пустая то возвращаем пустую строку
    if not s:
        return ""
    compressed = [] #список для хранения сжатых фрагментов
    count = 1 #счетчик одинаковых символов подряд

    # Проходим по строке, начиная со второго символа
    for i in range(1, len(s)):
        # Если текущий символ равен предыдущему
        if s[i] == s[i-1]:
            count += 1  #увеличиваем счетчик
        else:
            # Добавляем фрагмент: символ и количество 
            compressed.append(s[i-1] + (str(count) if count > 1 else  ""))
            count = 1  # Сбрасываем счетчик для нового символа
    # Добавляем последний фрагмент 
    compressed.append(s[-1] + (str(count) if count > 1 else ""))
    
    # Собираем все фрагменты в одну строку
    result = "".join(compressed)
    
    # Возвращаем сжатую версию, если она короче или равна исходной
    return result if len(result) <= len(s) else s
```

Задание 3 (10 баллов)
Проверка скобок

```python
def brackets(s):
    # Соответствие закрывающих скобок открывающим
    pairs = {')': '(', ']': '[', '}': '{'}
    # Стек для отслеживания открывающих скобок
    stack = []
    #Проходим по всем символам строки
    for char in s:
        # Если символ это открывающая скобка, добавляем в стек
        if char in '([{':
            stack.append(char)
        # Если символ это закрывающая скобка
        elif char in ')]}':
            # Проверяем, стек не пуст и верхний элемент соответствует
            if not stack or stack[-1] != pairs[char]:
                return False  #лишняя или непарная скобка
            # Удаляем соответствующую открывающую скобку
            stack.pop()
    return not stack
```

Задание 4 (10 баллов)
Генератор паролей

```python
import random
def gen_pass(n):
    # Проверяем на минимальную длину
    if n < 4: 
        return "Длина пароля должна быть 4 и больше символов"
    
    # Создаём строку со всеми возможными символами
    chars = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789!@#$%^&*_+-="
    
    #Гарантируем наличие всех типов символов:
    p = [
        random.choice("abcdefghijklmnopqrstuvwxyz"),  # случайная строчная буква
        random.choice("ABCDEFGHIJKLMNOPQRSTUVWXYZ"),  # случайная заглавная буква
        random.choice("0123456789"),                  # случайная цифра
        random.choice("!@#$%^&*_+-=")                # случайный спецсимвол
    ]
    #Добавляем остальные символы
    p += [random.choice(chars) for _ in range(n-4)]
    # Перемешиваем список символов,чтобы гарантированные символы не были в начале
    random.shuffle(p)
    
    # Превращаем список в строку
    return ''.join(p)
#Получаем длину от пользователя
try:
    # input() запрашивает ввод, если пользователь нажмёт Enter, будет использовано 12
    l = int(input("Length: ") or 12)
    # Генерируем и выводим пароль
    print(gen_pass(l))
except:
    print(gen_pass())
```

Задание 5 (6 баллов)
Почти палиндром

```python
def almost_palindrome(s):
    # Убираем пробелы и приводим к нижнему регистру
    s = s.lower().replace(" ", "")
    # Создаём функцию для проверки палиндрома
    def is_pal(x):
        return x == x[::-1]
    # Если уже палиндром возвращаем True
    if is_pal(s):
        return True, "палиндром"
    # Проверяем "почти палиндром"
    l, r = 0, len(s) - 1
    while l < r:
        if s[l] != s[r]:
            #Удаляем символ слева или справа
            left_removed = s[l+1:r+1]
            right_removed = s[l:r]
            # Проверяем, станет ли строка палиндромом после удаления
            if is_pal(left_removed) or is_pal(right_removed):
                return True, "почти палиндром"
            return False, "не палиндром"
        l += 1
        r -= 1
    return True, "палиндром"
```

Задание 6 (8 баллов)
Задача Шредингера

```python
import random 

def erase_simple(file, pct):
    # Открываем файл для чтения
    with open(file, 'r') as f:
        text = f.read()  
    # Разбиваем текст на слова и символы
    words = []  # Список для хранения слов и знаков препинания
    w = ""      # Временная переменная для накопления текущего слова
    # Проходим по каждому символу в тексте
    for c in text:
        # Если символ это буква, цифра, апостроф или дефис
        if c.isalnum() or c in "'-":
            w += c  #Добавляем к текущему слову
        else:
            # Если накопилось слово добавляем его в список
            if w: 
                words.append(w)  # Добавляем слово
                w = ""           # Сбрасываем временную переменную
            words.append(c)      # Добавляем текущий символ 
    
    # Добавляем последнее слово, если оно осталось
    if w: 
        words.append(w)
    
    #Выбираем случайные слова для удаления
    
    # Создаем список индексов реальных слов
    word_list = [i for i, w in enumerate(words) if w.strip()]
    
    # Вычисляем сколько слов нужно удалить 
    to_erase = random.sample(word_list, max(1, len(word_list) * pct // 100))
    
    #Заменяем выбранные слова на многоточие "..."
    for i in to_erase:
        words[i] = "..."  # Заменяем слово по индексу на "..."
    
    #Сохраняем результат в новый файл
    with open(f"del_{file}", 'w') as f:
        f.write("".join(words))  # Объединяем список в строку и записываем
#   Удаляет примерно 30% слов из файла "text.txt"
erase_simple("input.txt", 30)
```
